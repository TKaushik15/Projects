```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model
```

```python
df['diagnosis'] = pd.get_dummies(df.diagnosis)['M']
df
```

```python
x=df
```

```python
y=df.pop('diagnosis')
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y ,random_state = 123)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
mat=metrics.confusion_matrix(y_test,model.predict(x_test))
print(mat)
```

```python
from sklearn.metrics import accuracy_score
print("Train accuracy score: ",accuracy_score(y_test,model.predict(x_test)))
print("Test accuracy score: ",accuracy_score(y_train,model.predict(x_train)))
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
logistic_roc_auc = roc_auc_score(y_test,model.predict(x_test))
logistic_roc_auc
```

```python
fpr , tpr, thresholds = roc_curve(y_test,model.predict(x_test))
```

```python
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % logistic_roc_auc)
plt.plot([0, 1], [0, 1], '--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
